```python
import pandas as pd
import matplotlib.pyplot as plt

# Load each dataset
data_cluster = pd.read_csv("E:\\D\\HOCDIBANTRE\\DVRP-hwng\\data\\100\\h100c101.csv")
data_random = pd.read_csv("E:\\D\\HOCDIBANTRE\\DVRP-hwng\\data\\100\\h100r101.csv")
data_combined = pd.read_csv("E:\\D\\HOCDIBANTRE\\DVRP-hwng\\data\\100\\h100rc101.csv")

# Set up a figure with 3 subplots (1 row, 3 columns)
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Plot each dataset in a different subplot
axs[0].scatter(data_cluster['x'], data_cluster['y'], color='blue')
axs[0].set_title("Cluster Customer", fontsize=20)

axs[1].scatter(data_random['x'], data_random['y'], color='blue')
axs[1].set_title("Random Customer", fontsize=20)

axs[2].scatter(data_combined['x'], data_combined['y'], color='blue')
axs[2].set_title("Combined Cluster and Random Customer", fontsize=20)

# Adjust layout and show the plot
plt.tight_layout()
fig.savefig("plot_output.pdf", format='pdf', bbox_inches='tight')
plt.show()
```
