```python
import pandas as pd
import numpy as np

data_heuristic = r"Heuristic_result.csv"
data_gp = r"gp_result.csv"


result_heuristic = pd.read_csv(data_heuristic)
result_gp = pd.read_csv(data_gp)
result_heuristic.columns = result_heuristic.columns.str.strip()
result_gp.columns = result_gp.columns.str.strip()

data_frame = {
    "data_path":[],
    "num_vehicle":[],
    "fgp": [],
    "fheu1": [],
    "fheu2": [],
    "fheu3": [],
    "fheu4": [],
    "fheu5": [],
    "fheu6": [],
}

for index, row in result_heuristic.iterrows():
    for index_gp, row_gp in result_gp.iterrows():
        if row["data_path"] == row_gp["data_path"] and row["num_vehicle"] == row_gp["num_vehicle"]:
            data_frame["data_path"].append(row["data_path"])
            data_frame["num_vehicle"].append(row["num_vehicle"])
            data_frame["fgp"].append(row_gp["fgp"])
            data_frame["fheu1"].append(row["fheu1"])
            data_frame["fheu2"].append(row["fheu2"])
            data_frame["fheu3"].append(row["fheu3"])
            data_frame["fheu4"].append(row["fheu4"])
            data_frame["fheu5"].append(row["fheu5"])
            data_frame["fheu6"].append(row["fheu6"])
```

```python
result_store = r"merge.csv"
df = pd.DataFrame(data_frame)
df.to_csv(result_store, index=False)
```

```python
import matplotlib.pyplot as plt

result_data = r"merge.csv"
result = pd.read_csv(result_data)

iheu1 = result["iheu1"]
iheu2 = result["iheu2"]
iheu3 = result["iheu3"]
iheu4 = result["iheu4"]
iheu5 = result["iheu5"]
iheu6 = result["iheu6"]

# boxplot
plt.figure(figsize=(8, 6))
plt.boxplot([iheu1, iheu2, iheu3, iheu4, iheu5, iheu6], labels=["heu1", "heu2", "heu3", "heu4", "heu5", "heu6"])
plt.xlabel("Heuristic")
plt.ylabel("Improvement (%)")
```
